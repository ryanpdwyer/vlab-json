```python
# Use a json library that is less strict about the format
import json5 as json
data_dict = json.loads(data)

# HTML print the reactions
# Okay for it to only work in IPython notebook
from IPython.display import HTML
```

```python
# Calculate Delta H, Delta S, and Delta G for each reaction

# Get the reactions
reactions = data_dict['reactions']['REACTIONS']['REACTION']
# Get the species
species = data_dict['species']['SPECIES_LIST']['SPECIES']
# Turn the list of species with IDs to a dictionary with id as the key
species_dict = {s['id']: s for s in species}

def string_reaction_side(side):
    species_strings = []
    for species in side:
        if species['coeff'] == 1:
            species_strings.append(species['name'])
        else:
            species_strings.append(str(species['coeff']) + species['name'])
    return ' + '.join(species_strings)

# Calculate the delta H, delta S, and delta G for each reaction
for reaction in reactions:
    # Get the species in the reaction
    species_in_reaction = reaction['SPECIES_REF']
    # Calculate the delta H
    delta_H = sum([float(species_dict[s['id']]['enthalpy']) * float(s['coefficient']) for s in species_in_reaction])
    # Calculate the delta S
    delta_S = sum([float(species_dict[s['id']]['entropy']) * float(s['coefficient']) for s in species_in_reaction])
    # Calculate the delta G
    delta_G = delta_H - 298.15 * delta_S / 1000
    # Add the delta H, delta S, and delta G to the reaction
    reaction['delta_H'] = delta_H
    reaction['delta_S'] = delta_S
    reaction['delta_G'] = delta_G
    reaction['log_K'] = -delta_G / (2.303 * 298.15 * 8.3145)
    # Write the reaction to a string
    reactants = []
    products = []
    for species_ in species_in_reaction:
        # Convert to an int if the coefficient is an integer
        if float(species_['coefficient']) == int(species_['coefficient']):
            coeff = int(species_['coefficient'])
        else:
            coeff = float(species_['coefficient'])
        if float(species_['coefficient']) < 0:
            reactants.append(dict(name=species_dict[species_['id']]['name'], coeff=-coeff))
        else:
            products.append(dict(name=species_dict[species_['id']]['name'], coeff=coeff))

    reaction_string = string_reaction_side(reactants) + ' -> ' + string_reaction_side(products)
    reaction['name'] = reaction_string
    reaction['html'] = HTML(reaction_string)
```
